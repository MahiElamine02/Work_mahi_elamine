```python
#import packages
import pandas as pd
import matplotlib.pyplot as plt

#import fichier 2022
#skiprows élimine les trois premières lignes à l'import
chemin = r"C:\Users\leogu\OneDrive\Bureau\Etude\Master\FinanceIndustrie\projet\donnees\ownershiptotal2022.xlsx" #créer le chemin
data_2022 = pd.read_excel(chemin, skiprows=3) #importer donnees csv via le chemin #pour xlsx : pd.read_excel 
```

```python
#total action 2022
total_shares_2022 = data_2022['Investor Shares Held'].sum()

#% action par actionnaire arrondi au centième
data_2022['Percentage of Total Shares_2022'] = (data_2022['Investor Shares Held'] / total_shares_2022) * 100
data_2022['Percentage of Total Shares_2022'] = data_2022['Percentage of Total Shares_2022'].round(1)

#tri action 2022 ordre décroissant
data_sorted_2022 = data_2022.sort_values(by='Investor Shares Held', ascending=False)

#colonnes 10 premiers investisseurs
top_investors_2022 = data_sorted_2022[['Investor Full Name', 'Investor Shares Held', 'Percentage of Total Shares_2022']].head(20)

#renommer colonnes en français
top_investors_2022 = top_investors_2022.rename(columns={"Investor Full Name": "Nom de l'investisseur_2022", 
                                                        "Investor Shares Held": "Actions détenues_2022",
                                                        "Percentage of Total Shares_2022":"% Total des actions_2022"})

# % 20 premiers
investor_names_2022 = top_investors_2022["Nom de l'investisseur_2022"]
percentages_2022 = top_investors_2022["% Total des actions_2022"]
restant_percentages_2022 = 100 - percentages_2022.sum()
restant_invest_2022 = len(data_2022["Investor Full Name"]) - 20
restant_percentages_2022 = restant_percentages_2022.round(1)

#graph barres histo
plt.figure(figsize=(12, 8))
bars = plt.bar(investor_names_2022, percentages_2022, color='skyblue', width=0.6)  # Ajustement de la largeur des barres

#% au-dessus des barres
for bar, percentage in zip(bars, percentages_2022):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), f'{percentage}%', ha='center', va='bottom')

#titres
plt.title('Répartition des actions entre les 20 premiers investisseurs pour l\'année 2022 (supérieur à 1%)')
plt.xlabel('Investisseurs')
plt.ylabel('% des actions détenues')

#affichage étiquette axe X sur le coté
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

#texte
print("De plus, il est important de noter que", restant_percentages_2022, "% est détenu par",
      restant_invest_2022,"autres investisseurs. On a donc un total de", len(data_2022["Investor Full Name"]), "investisseurs.")
print("Le nombre total d'action est de",total_shares_2022,"dont 13.5% détenu par Total Employees, qui sont les premiers investisseurs.")


```

```python
#Calcul hhi 2022
calc_hhi_2022 = top_investors_2022.head(5)
hhi_2022 = (calc_hhi_2022['% Total des actions_2022'] ** 2).sum()
print("Indice de Herfindahl-Hirschman (HHI) :", hhi_2022)
```
